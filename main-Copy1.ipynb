```python
import import_ipynb
from Emotion_Detection import run_emotion_detection
from music_recommendation import create_spotify_playlist

def main():
    print("Starting Emotion Detection...")
    detected_emotion = run_emotion_detection()
    print(f"Detected Emotion: {detected_emotion}")

    user_choice = input(f"You seem {detected_emotion}. Would you like me to create a playlist for you? (yes/no): ").strip().lower()
    if user_choice == "yes":
        playlist_name = create_spotify_playlist(detected_emotion)
        print(f"Your playlist '{playlist_name}' has been created! Enjoy!")
    else:
        print("No playlist created. Have a nice day!")

if __name__ == "__main__":
    main()
```

```python
import import_ipynb
from Emotion_Detection import run_emotion_detection
from music_recommendation import create_spotify_playlist
import ttkbootstrap as ttk
from ttkbootstrap.constants import *
from tkinter import messagebox


class EmotionChatbotApp(ttk.Window):
    def __init__(self):
        super().__init__(themename="cyborg")  # Modern theme
        self.title("Emotion Detection Chatbot")
        self.geometry("600x700")
        self.resizable(False, False)

        # Initialize variables
        self.detected_emotion = None
        self.chat_history = []
        self.chat_stage = "permissions"
        self.chat_prompt_index = 0  # Track the index for emotion-specific prompts
        self.permissions_granted = False  # Track if permissions are granted
        self.playlist_name = None  # Track the playlist name created

        # Emotion-specific prompts
        self.prompts = {
            "Happy": [
                "How are you feeling today?",
                "Would you like an upbeat playlist?",
                "Do you have a favorite artist in mind?",
                "Shall we explore trending tracks for a good vibe?"
            ],
            "Sad": [
                "Would you like a calming playlist?",
                "Do you want to talk about what’s making you feel this way?",
                "How about some acoustic music to lift your mood?",
                "Would you like me to create a motivational playlist?"
            ],
            "Stressed": [
                "Feeling stressed? Want a relaxing playlist?",
                "How about some ambient music to help you unwind?",
                "Would you like to talk about your day?",
                "How about a short mindfulness activity to calm down?"
            ],
            "Bored": [
                "Feeling bored? Let’s explore some exciting tracks.",
                "Would you like to discover new genres?",
                "How about an upbeat playlist to energize you?",
                "Do you have a specific song in mind to add?"
            ]
        }

        # General prompts
        self.general_prompts = [
            "What’s your favorite song right now?",
            "Would you like me to recommend a podcast?",
            "Shall we explore some trending genres?",
            "Would you like to know fun facts about music?"
        ]

        # Emotion to genre mapping (for prompts and suggestions, not playlist creation)
        self.emotion_to_genre = {
            "Happy": "pop",
            "Sad": "acoustic",
            "Neutral": "indie",
            "Angry": "rock",
            "Relaxed": "chill",
            "Stressed": "ambient",
            "Bored": "classical"
        }

        # Create UI elements
        self.create_widgets()

    def create_widgets(self):
        # Header
        header_frame = ttk.Frame(self, padding=10)
        header_frame.pack(fill=X)

        header_label = ttk.Label(header_frame, text="Emotion Detection Chatbot", font=("Helvetica", 20, "bold"))
        header_label.pack()

        # Chat Area
        self.chat_frame = ttk.Frame(self, padding=20)
        self.chat_frame.pack(fill=BOTH, expand=True)

        self.chat_display = ttk.Text(self.chat_frame, wrap=WORD, state=DISABLED, font=("Helvetica", 12))
        self.chat_display.pack(fill=BOTH, expand=True)

        # Input and Buttons
        input_frame = ttk.Frame(self, padding=10)
        input_frame.pack(fill=X)

        self.user_input = ttk.Entry(input_frame, font=("Helvetica", 14))
        self.user_input.pack(side=LEFT, fill=X, expand=True, padx=5)

        send_button = ttk.Button(input_frame, text="Send", command=self.handle_user_input, bootstyle=PRIMARY)
        send_button.pack(side=RIGHT)

        # Start Button
        self.start_button = ttk.Button(self, text="Start Emotion Detection", command=self.ask_permissions, bootstyle=SUCCESS)
        self.start_button.pack(pady=10)

    def add_chat_message(self, message, sender="Chatbot"):
        """
        Add a message to the chat area with styled chat bubbles.
        """
        self.chat_display.config(state=NORMAL)
        self.chat_display.insert("end", f"{sender}: {message}\n", "chatbot" if sender == "Chatbot" else "user")
        self.chat_display.see("end")
        self.chat_display.config(state=DISABLED)

    def ask_permissions(self):
        """
        Ask for the required permissions using Yes/No dialogs.
        """
        if self.permissions_granted:
            return  # If permissions were already granted, don't ask again
        
        # Step 1: Asking for permissions
        if not messagebox.askyesno("Permission Request", "Do you allow us to capture a photo to detect your facial emotions?"):
            self.add_chat_message("Permission for photo capture denied. Exiting journey.", sender="Chatbot")
            return

        # Wearable device permission
        if not messagebox.askyesno("Permission Request", "Do you allow us to access your ECG and respiratory levels from your wearable device?"):
            self.add_chat_message("Permission for wearable access denied. Exiting journey.", sender="Chatbot")
            return

        # Sleep pattern permission
        if not messagebox.askyesno("Permission Request", "Do you allow us to access your sleep pattern data?"):
            self.add_chat_message("Permission for sleep data denied. Exiting journey.", sender="Chatbot")
            return

        self.permissions_granted = True
        self.add_chat_message("Permissions granted! Starting Emotion Detection...", sender="Chatbot")
        self.start_emotion_detection()

    def start_emotion_detection(self):
        """
        Run emotion detection and proceed with the chatbot flow.
        """
        # Only run emotion detection once
        if self.detected_emotion:
            return

        self.detected_emotion = run_emotion_detection()
        self.add_chat_message(f"Your detected emotion is: {self.detected_emotion}", sender="Chatbot")

        # Initialize the chat stage and start prompts based on the detected emotion
        self.chat_stage = "emotion_prompt"
        self.show_next_prompt()

    def show_next_prompt(self):
        """
        Display the next prompt based on the chat stage.
        """
        if self.chat_stage == "emotion_prompt":
            # Emotion-specific prompts
            emotion_prompts = self.prompts.get(self.detected_emotion, [])
            if emotion_prompts:
                if self.chat_prompt_index < len(emotion_prompts):
                    prompt = emotion_prompts[self.chat_prompt_index]
                    self.add_chat_message(prompt)
                    self.chat_prompt_index += 1
                else:
                    self.chat_stage = "general_prompt"
                    self.show_next_prompt()
            else:
                self.chat_stage = "general_prompt"
                self.show_next_prompt()

        elif self.chat_stage == "general_prompt":
            if self.general_prompts:
                prompt = self.general_prompts.pop(0)
                self.add_chat_message(prompt)
            else:
                self.add_chat_message("I’m here to help. Let me know what you’d like to explore!")

    def handle_user_input(self):
        """
        Handle user input from the text box.
        """
        user_message = self.user_input.get().strip()
        if not user_message:
            return

        # Display user message
        self.add_chat_message(user_message, sender="You")
        self.user_input.delete(0, "end")

        # Process user response based on keywords or phrases
        user_message = user_message.lower()

        if "create" in user_message and "playlist" in user_message:
            # If the user mentions a specific genre (ignoring the emotion-based genre)
            if "chill" in user_message or "relax" in user_message:
                self.create_playlist("chill")
            elif "upbeat" in user_message or "energetic" in user_message:
                self.create_playlist("upbeat")
            elif "rock" in user_message:
                self.create_playlist("rock")
            elif "pop" in user_message:
                self.create_playlist("pop")
            elif "classical" in user_message:
                self.create_playlist("classical")
            elif "ambient" in user_message:
                self.create_playlist("ambient")
            else:
                self.add_chat_message("Sorry, I didn’t recognize that genre. Please choose from: chill, pop, rock, ambient, etc.", sender="Chatbot")
        elif "change" in user_message and "playlist" in user_message:
            self.add_chat_message("Sure, let’s create a new playlist for you! What kind would you like?", sender="Chatbot")
            self.chat_stage = "emotion_prompt"  # Allow new playlist creation
            self.chat_prompt_index = 0  # Reset prompt index
            self.playlist_name = None  # Clear previous playlist name
        elif "suggest" in user_message or "play" in user_message:
            self.add_chat_message("Let me suggest some songs for your mood!", sender="Chatbot")
        elif "trending" in user_message:
            self.add_chat_message("Here are some trending songs for you!", sender="Chatbot")
        elif "no" in user_message:
            self.add_chat_message("No problem! Let me know if you need anything else.", sender="Chatbot")

    def create_playlist(self, genre):
        """
        Create a Spotify playlist based on the genre.
        """
        if genre not in ["chill", "pop", "rock", "ambient", "classical"]:
            self.add_chat_message("Sorry, I didn’t recognize that genre. Let me try again.", sender="Chatbot")
            return
        
        # Always create a playlist with the requested genre
        self.playlist_name = create_spotify_playlist(genre)
        self.add_chat_message(f"Here’s your {genre} playlist: {self.playlist_name}", sender="Chatbot")


if __name__ == "__main__":
    app = EmotionChatbotApp()
    app.mainloop()
```
